```python
import pandas as pd

df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')

df.head()
```

```python
# Step 3: Basic Info
print("Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nInfo:\n")
df.info()
```

```python
# Step 4: Summary Statistics
df.describe()
```

```python
#Step 5: Select Columns
# Example columns
df[['Sales', 'Profit', 'Category']].head()
```

```python
#Step 6: Filtering Data# High sales
high_sales = df[df['Sales'] > 500]
high_sales.head()

# Loss-making orders
loss = df[df['Profit'] < 0]
loss.head()
```

```python
#Step 7: Grouping
# Total sales by category
df.groupby('Category')['Sales'].sum()

# Profit by region
df.groupby('Region')['Profit'].sum()
```

```python
#Step 8: Top Records
# Top 5 highest sales
df.sort_values(by='Sales', ascending=False).head()

# Top 5 profit
df.sort_values(by='Profit', ascending=False).head()
```

```python
#Step 9: Save Filtered Data
high_sales.to_csv('high_sales.csv', index=False)
```

```python
#more
#Find Best Selling Category
best_category = df.groupby('Category')['Sales'].sum().idxmax()
print("Best Category:", best_category)
```

```python
#Region with Highest Profit
best_region = df.groupby('Region')['Profit'].sum().idxmax()
print("Best Region:", best_region)
```
