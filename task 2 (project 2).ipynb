```python
# Core libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
sns.set(style="whitegrid")
```

```python
import seaborn as sns

df = sns.load_dataset("tips")

df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['total_bill'], kde=True, bins=20)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.savefig("histogram_kde.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['total_bill'])

plt.title("Boxplot of Total Bill")
plt.savefig("boxplot.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='total_bill', hue='time', kde=True)

plt.title("Distribution by Time (Lunch vs Dinner)")
plt.savefig("group_comparison.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='day', y='total_bill', data=df)

plt.title("Total Bill by Day")
plt.savefig("boxplot_by_day.png")
plt.show()
```

```python
Q1 = df['total_bill'].quantile(0.25)
Q3 = df['total_bill'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['total_bill'] < Q1 - 1.5*IQR) | (df['total_bill'] > Q3 + 1.5*IQR)]

print("Number of outliers:", len(outliers))
outliers.head()
```

```python
skewness = df['total_bill'].skew()
print("Skewness:", skewness)
```

The distribution of total bill values shows a right-skewed pattern, indicating that most customers spend lower amounts while a few spend significantly higher amounts. The histogram and KDE plot confirm this positive skewness. Boxplot analysis reveals the presence of outliers on the higher end, suggesting occasional high-value transactions. When comparing groups, dinner transactions tend to have higher variability and larger bills than lunch. Overall, the data exhibits moderate spread with noticeable outliers and a non-normal distribution.
